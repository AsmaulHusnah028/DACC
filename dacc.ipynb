##**3. Menghilangkan teks link dari column tweet**

```python
import re
```

```python
tweet_noxspace = []
for tw in df_concat['tweet']:
  tweet_noxspace.append(" ".join(tw.split()))
```

```python
tweet_nonumber = []
# Remove numbers
for tw in tweet_noxspace:
  tweet_nonumber.append(re.sub(r'\w*\d\w*', '', tw) )
```

```python
tweet_nohashtag = []
for tw in tweet_nonumber:
  tweet_nohashtag.append(re.sub("#[A-Za-z0-9_]+","", tw))
```

```python

tweet_nolink= []
for tw in tweet_nohashtag:
  tweet_nolink.append(re.sub(r'http\S+', '',  tw))
tweet_nolink
```

##**4. Stemming Word**

```python
!pip install git+https://github.com/har07/PySastrawi.git
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

```python
type(tweet_nolink)
```

```python
# create stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

tweet_stemmed = []
for tw in tweet_nolink:
  tweet_stemmed.append(stemmer.stem(tw))
```

```python
type(tweet_stemmed)
```
